# 2 (e) treating the outliers using imputing method

```python
q1 = df['Sale Price'].quantile(0.25)
q3 = df['Sale Price'].quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
lower_limit = q1 - 1.5*IQR
upper_limit = q3 + 1.5*IQR
lower_limit , upper_limit
```

```python
# bcz upper limit is < min value of sale price
lower_limit = df['Sale Price'].min()
lower_limit
```

```python
def outlier_remover(value):
    if value > upper_limit :
        return upper_limit
    if value < lower_limit :
        return lower_limit
    else :
        return value 
```

```python
df['Sale Price'] = df['Sale Price'].apply(outlier_remover)
```

# Treating Multicollinearity

```python
Y = df['Sale_Price']
X = df.drop(columns = 'Sale_Price')
X.shape
```

```python
X.corr()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X = pd.DataFrame(data = X , columns = df.drop(columns = ['Sale_Price']).columns)
```

```python
VIF = pd.Series([variance_inflation_factor(X.values , i) for i in range(X.shape[1])] , index = X.columns)
VIF
```

```python
def MC_remover(data):
    vif = pd.Series([variance_inflation_factor(data.values , i)  for i in range (data.shape[1])] ,index = data.columns)
    
    if vif.max() > 5 :
        print(vif[vif==vif.max()].index[0] , 'has been removed')
        data = data.drop(columns = [vif[vif==vif.max()].index[0]])

        return data
    else :
        print('no multicollinearity present anymore')
        return data
    
    
    
    
for i in range(7):
    X = MC_remover(X)
    
```
